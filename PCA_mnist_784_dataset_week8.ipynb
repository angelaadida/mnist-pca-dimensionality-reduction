# Principal Component Analysis (PCA) on mnist_784 Dataset


This PCA on the mnist_784 dataset shows how the method captures the most prominent patterns in handwritten digits. 
The visualized principal components (eigen-digits) reflect essential features differentiating the digits.

## 1. Importing Necessary Libraries

```python
# Importing required libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_openml
from sklearn.decomposition import PCA
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

## 2. Loading and Visualizing the Dataset

```python
 # Loading the mnist_784 dataset
mnist = fetch_openml('mnist_784', version=1)
digits = mnist.data / 255.0  

# Visualizing the first digit from the dataset
plt.imshow(digits.values[0].reshape(28, 28), cmap='gray')
plt.title('Sample Digit')
plt.axis('off')
plt.show()
```

## 3. Applying PCA

```python
# Using PCA to reduce dimensionality
n_components = 100 
pca = PCA(n_components=n_components, whiten=True).fit(digits)

# Transforming the digit data using the PCA
digits_pca = pca.transform(digits)
```

## 4. Visualizing PCA Results

```python
eigen_digits = pca.components_.reshape((n_components, 28, 28))
plt.figure(figsize=(10, 10))
for i in range(25): 
    plt.subplot(5, 5, i + 1)
    plt.imshow(eigen_digits[i], cmap='gray')
    plt.title(f"PC {i+1}")
    plt.axis('off')
plt.tight_layout()
plt.show()
```
